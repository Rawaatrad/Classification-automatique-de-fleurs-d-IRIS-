```python
#importation des bibliothéques
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_scoreimport matplotlib.pyplot as plt
```

```python
#chargement des données
iris = load_iris()
X, y = iris.data, iris.target
```

```python
#séparation des données 70% pour l’apprentissage, 30% pour le test.
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=0)
```

```python
#construction de l'arbre de décision sur les données IRIS
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
fn = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
cn = ['setosa', 'versicolor', 'virginica']
```

```python
#visualisation de l'arbre
plt.figure(figsize = (10,8))
tree.plot_tree(clf, feature_names = fn, class_names = cn, filled = True);
```

```python
clf.predict(X_test)
```

```python
#construction de l'arbre de décision avec le paramétre max_depth
clf = tree.DecisionTreeClassifier(max_depth=3)
clf.fit(X_train, y_train)
```

```python
#visualisation de l'arbre
plt.figure(figsize = (10,8))
tree.plot_tree(clf, feature_names = fn, class_names = cn, filled = True);
```

```python
clf.predict(X_test)
```

```python
#construction de l'arbre de décision avec le paramétre min_sample_leaf
clf = tree.DecisionTreeClassifier(min_samples_leaf=20)
clf.fit(X_train, y_train)
```

```python
#visualisation de l'arbre
plt.figure(figsize = (10,8))
tree.plot_tree(clf, feature_names = fn, class_names = cn, filled = True);
```

```python
clf.predict(X_test)
```
